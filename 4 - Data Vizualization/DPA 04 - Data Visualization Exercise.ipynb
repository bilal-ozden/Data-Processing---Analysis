What are you trying to show, and how can you use visual design to tell that story effectively? Keep the following design elements in mind (not all apply to every plot):

- **Position:** The position is determined by the plot type (e.g., `plot(x, y)`, `scatter(x, y)`, `hist(x)`)
- **Shape:** You can use the `marker` option in matplotlib. Note: you can only use one marker per plot. Can you think of a workaround?
- **Size:** You can use the `s` option in matplotlib
- **Color:** You can use the `c` or `color` option (`color` is preferred for a single color; `c` for an array)
- **Line width:** You can use the `linewidth` option in matplotlib
- **Line type:** You can use the `linestyle` option in matplotlib
