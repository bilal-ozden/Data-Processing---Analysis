By applying .describe() to the boolean DataFrame from .isna(), we get counts of True (missing) and False (present) values per column, helping us quantify missing data.

- **count:** The total number of rows in the dataset (e.g., 892 for the Titanic dataset). This is the same for every column since .describe() is applied to the full DataFrame.

- **unique:** The number of unique values in the column. For boolean data (True/False), this is usually 2 (True and False), but it could be 1 if a column has no missing values (only False) or is entirely missing (only True).

- **top:** The most common value in the column. For missing data analysis, this is often False (meaning most values are present) or True (meaning most values are missing).

- **freq:** The frequency (count) of the top value. This tells you how many rows have the most common state (e.g., how many are missing or present).

## 3.3 First data transformations 

Now that we've discussed modifying data types, let's start doing it!

- The easiest and most common way is called **one-hot encoding**, also known as **dummy variables**. 

- The idea is simple: for a binary variable (either A or B), we can represent it with the numbers 0 and 1.

- This column represents **whether the variable is present**. It's important to decide what "present" means for our interpretation.

- In the Titanic dataset, a good candidate is the 'Survived' column. We can define survival as the variable being present, so 'Yes' becomes 1 and 'No' becomes 0.

How do we implement this? Manually replacing each value would take way too long. Fortunately, pandas has the .map() function built-in.

We call the column we want to map and pass a dictionary where the keys are the old values and the values are the new ones.

```python
titanic['Survived'] = titanic['Survived'].map({'No': 0, 'Yes' : 1})
```

The .map() method applies a dictionary mapping to replace values in the Series. Here, it converts 'No' to 0 and 'Yes' to 1.

```python
titanic['Survived']
```

### Question 4: One-hot encoding works by indicating whether a certain variable is present, thus limiting it to binary columns only. This means we cannot simply apply it to columns like 'Embarked'. Can you think of a way to use this dummy variables technique anyways to encode such columns? Are there any new limitations for this?

A column like 'Embarked' has multiple possible values. While we cannot convert it to a single column indicating presence, we can create **a new column for each possible value** and mark if that value is present for each record. We must ensure this transformation makes sense.

For example, in a car dataset, applying this to the model name column would not add value, since each car has a unique model. Each new column would have mostly zeros and one 1, increasing the dataset size without benefit. We'll discuss this more in the Introduction to Machine Learning class. However, it could be useful for the brand name column, where many cars share the same brand, providing real information gain.

Pandas provides the .get_dummies() function for this.

```python
titanic = pd.get_dummies(titanic, columns = ['Embarked'])
```

pd.get_dummies() creates new binary columns for each unique value in the specified column(s). Here, it adds columns like 'Embarked_C', 'Embarked_Q', 'Embarked_S' with 1 where the value matches and 0 otherwise.

```python
titanic.sample(5)
```

Great! But are these the only ways to transform data? Of course not! Sometimes datasets already use booleans for binary data. Pandas makes it easy to convert data types, as long as they are compatible (like True/False and 1/0).

```python
titanic['Survived'].astype('bool')
```

The **.astype('bool')** method converts the Series to boolean type:
-  changing 0 to -> False
-  1 to -> True.

## 3.4 Missing values

There are three main ways to handle missing data:

- Removing data

- Data imputation

- Data flagging

Removing data is the simplest, but it leads to the most information loss.

```python
titanic2 = titanic.dropna()
```

dropna() removes any row containing at least one missing value, which can drastically reduce the dataset size.

```python
titanic.isna().describe()
```

```python
titanic2.isna().describe()
```

Notice how easy it was to drop all missing data, but it left us with only 185 rows out of 891 – a huge loss! By default, .dropna() deletes any row with a missing value. To adjust this, we can set a threshold.

```python
titanic2 = titanic.dropna(thresh=5)
```

dropna(thresh=5) keeps only rows with at least 5 non-missing values, providing more control over data removal.

```python
titanic2.isna().describe()
```

Now we haven't removed any rows with missing values. You can see how the threshold can be tricky to tune!

Another way to handle missing values is by filling them with reasonable estimates. There are several built-in functions for this!

```python
titanic['Age'].value_counts()
```

```python
age = titanic['Age'].ffill()
age.value_counts()
```

ffill() fills missing values with the last known (previous) value in the column.

```python
age = titanic['Age'].bfill()
age.value_counts()
```

bfill() fills missing values with the next known (subsequent) value in the column.

Sometimes built-in functions aren't suitable, so you need a custom approach.

```python
age = titanic['Age'].fillna(titanic['Age'].median())
age.value_counts()
```

fillna() with the median fills missing values with the middle value of the non-missing data, preserving the central tendency.

Remember to be very careful when filling data like this! It's easy for a value to seem good but actually distort the dataset. That's why sometimes it's better to 'flag' missing values by assigning a special value that doesn't interfere with the data.

Make sure to choose values that won't ruin your dataset – avoid extremes like very low or high numbers!

```python
age = titanic['Age'].fillna(0)
age.value_counts()
```

Filling with 0 flags missing values but might skew statistics if 0 is a valid age.

```python
titanic['Age_missing'] = titanic['Age'].isna().astype(int)
age = titanic['Age'].fillna(titanic['Age'].median())  # Or another imputation method
age
```
