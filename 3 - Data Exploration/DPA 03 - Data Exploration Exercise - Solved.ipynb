## 3.3 Data transformations

Transform the data types of your columns! Make sure your data is in the desired form so you can easily handle the missing values later.

*Hint: We have not yet seen what to do with data in columns like 'Name', 'Ticket', and 'Cabin'. Thus, you can leave those three columns as they are.*

Transform the data types of your columns! Make sure your data is in the desired from so you can easily handle the missing values later.

```python
titanic['Survived'] = titanic['Survived'].map({'No': 0, 'Yes' : 1})
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female' : 1})
titanic = pd.get_dummies(titanic, columns = ['Embarked'])
titanic['Embarked_C'] = titanic['Embarked_C'].astype('float')
titanic['Embarked_Q'] = titanic['Embarked_Q'].astype('float')
titanic['Embarked_S'] = titanic['Embarked_S'].astype('float')
```

The .map() method replaces values based on a dictionary. Here, it converts 'No'/'Yes' to 0/1 for 'Survived' and 'male'/'female' to 0/1 for 'Sex'.

pd.get_dummies() creates binary columns for each unique value in 'Embarked'.

.astype('float') converts the new dummy columns to float type for consistency.

Checking the data type of 'Fare' before cleaning.

```python
def clean_fare(value):
    if len(value) > 3:
        if value[-3] == '.' and value[-2:].isdigit():
            value = value.replace('.', ',', 1)
        value = value.replace('.', '')
        value = value.replace(',', '.')
    return value

titanic['Fare'] = titanic['Fare'].apply(clean_fare)
```

This custom function cleans the 'Fare' values by handling European number formatting (commas as decimals) and converting to standard float format.

### 3.4.2: Data Imputation

```python
titanic3 = titanic
```

Creating a copy of the DataFrame for imputation experiments.

```python
titanic3.isna().describe()
```

ffill() fills missing values with the last known value from previous rows.

```python
titanic3['Survived'] = titanic3['Survived'].ffill()
titanic3['Pclass'] = titanic3['Pclass'].ffill()
titanic3['Name'] = titanic3['Name'].ffill()
titanic3['Sex'] = titanic3['Sex'].ffill()
titanic3['Age'] = titanic3['Age'].ffill()
titanic3['SibSp'] = titanic3['SibSp'].ffill()
titanic3['Parch'] = titanic3['Parch'].ffill()
titanic3['Ticket'] = titanic3['Ticket'].ffill()
titanic3['Fare'] = titanic3['Fare'].ffill()
titanic3['Cabin'] = titanic3['Cabin'].ffill()
```

ffill() fills missing values with the last known value from previous rows.

```python
titanic3.isna().describe()
```

We can see that the 'Age' and 'Cabin' columns still have missing values. Let's investigate!

```python
titanic3.loc[np.logical_or(titanic3['Cabin'].isna(), titanic3['Age'].isna())]
```

This locates rows where 'Age' or 'Cabin' are still missing after ffill.

Aha, this makes sense! The first record has missing values, so there's no previous value to forward fill. We'll use bfill() to fill from subsequent values.

```python
titanic3['Age'] = titanic3['Age'].bfill()
titanic3['Cabin'] = titanic3['Cabin'].bfill()
titanic3.isna().describe()
```

bfill() fills missing values with the next known value from subsequent rows.

### 3.4.3: Data Flagging

Unfortunately, Python creates copies by reference, so modifying the 'copy' affects the original. We need to reload the data.

```python
titanic = pd.read_csv("titanic_uncleaned.csv", sep = ";")

titanic['Survived'] = titanic['Survived'].map({'No': 0, 'Yes' : 1})
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female' : 1})
titanic = pd.get_dummies(titanic, columns = ['Embarked'])
titanic['Embarked_C'] = titanic['Embarked_C'].astype('float')
titanic['Embarked_Q'] = titanic['Embarked_Q'].astype('float')
titanic['Embarked_S'] = titanic['Embarked_S'].astype('float')

def clean_fare(value):
    value = str(value)
    if len(value) > 3:
        if value[-3] == '.' and value[-2:].isdigit():
            value = value.replace('.', ',', 1)
        value = value.replace('.', '')
        value = value.replace(',', '.')
    return value

titanic['Fare'] = titanic['Fare'].apply(clean_fare).astype('float')
```

Reloading and reapplying transformations, then using fillna() to flag missing values with special numbers (-1 for numeric, "" for strings).

```python
titanic['Survived'] = titanic['Survived'].fillna(-1)
titanic['Pclass'] = titanic['Pclass'].fillna(-1)
titanic['Name'] = titanic['Name'].fillna("")
titanic['Age'] = titanic['Age'].fillna(-1)
titanic['SibSp'] = titanic['SibSp'].fillna(-1)
titanic['Parch'] = titanic['Parch'].fillna(-1)
titanic['Ticket'] = titanic['Ticket'].fillna("")
titanic['Fare'] = titanic['Fare'].fillna(-1)
titanic['Cabin'] = titanic['Cabin'].fillna("")
```

```python
titanic.isna().describe()
```

Confirming no missing values remain after flagging.

### 3.5: Custom approach

Combine the knowledge you've gained by comparing the three methods to create one final cleaned dataset where all missing values are handled optimally!

Combine the knowledge you have gained by comparing the three methods to create one final cleaned dataset where alle missing values are handled in the most optimal way!

```python
titanic = pd.read_csv("titanic_uncleaned.csv", sep = ";")

titanic['Survived'] = titanic['Survived'].map({'No': 0, 'Yes' : 1})
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female' : 1})
titanic = pd.get_dummies(titanic, columns = ['Embarked'])
titanic['Embarked_C'] = titanic['Embarked_C'].astype('float')
titanic['Embarked_Q'] = titanic['Embarked_Q'].astype('float')
titanic['Embarked_S'] = titanic['Embarked_S'].astype('float')

def clean_fare(value):
    value = str(value)
    if len(value) > 3:
        if value[-3] == '.' and value[-2:].isdigit():
            value = value.replace('.', ',', 1)
        value = value.replace('.', '')
        value = value.replace(',', '.')
    return value

titanic['Fare'] = titanic['Fare'].apply(clean_fare).astype('float')
```

Reloading data and applying a mixed approach: drop 'Cabin', use ffill for some, median for 'Age', etc.

```python
del titanic['Cabin']
```

Deleting the 'Cabin' column since it has too many missing values.

```python
titanic['Survived'] = titanic['Survived'].ffill()
titanic['Pclass'] = titanic['Pclass'].ffill()
titanic['Name'] = titanic['Name'].fillna("")
titanic['Sex'] = titanic['Sex'].ffill()
titanic['Age'] = titanic['Age'].fillna(titanic[titanic['Age'].notna()].Age.median())
titanic['SibSp'] = titanic['SibSp'].fillna(titanic[titanic['SibSp'].notna()].Age.median())
titanic['Parch'] = titanic['Parch'].fillna(titanic[titanic['Parch'].notna()].Age.median())
titanic['Ticket'] = titanic['Ticket'].fillna("")
titanic['Fare'] = titanic['Fare'].fillna(titanic[titanic['Fare'].notna()].Age.median())
```

Applying a mix of ffill, fillna with median, and empty strings for optimal handling.
